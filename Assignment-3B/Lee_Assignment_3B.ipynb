````python
## Code goes here

import pandas as pd

df = pd.read_csv("recweekdata.csv") # read
df2 = pd.read_csv("studentprofiles.csv") # read
````

````python
## birthmonth

bmDict = {"1":"January","2":"February","3":"March","4":"April","5":"May","6":"June",
        "7":"July","8":"August","9":"September","10":"October","11":"November","12":"December",
        "Jan":"January","Feb":"February","Mar":"March","Apr":"April","Jun":"June",
        "Jul":"July","Aug":"August","Sep":"September","Oct":"October","Nov":"November",
        "Dec":"December","January":"January","February":"February","March":"March","April":"April",
        "May":"May","June":"June","July":"July","August":"August","September":"September","October":"October",
        "November":"November","December":"December"} # dictionary to clean

df.loc[:,"birthmonth"] = df.loc[:,"birthmonth"].apply(lambda month: bmDict[month]) # subject all to Dict

df.head()
````

````python
## birthyear

import re

def yearCleaner(year):
    if year >= 1900 and year < 2000: # 19XX
        return re.sub("^19","",str(year)) # remove 19
    elif year >= 2000: # 20XX
        return re.sub("^20","",str(year)) # remove 20
    else:
        return str(year) # for XX, keep as is

df.loc[:,"birthyear"] = df.loc[:,"birthyear"].apply(yearCleaner) # apply to all

df.head()
````

````python
## view all possible responses for course

df.loc[:,["firstname","course"]].groupby(["course"]).count()
````

````python
## course

courseDict = {"BS CTM":"BS Communications Technology Management", "BS ITE":"BS Information Technology Entrepreneurship",
             "BS LM":"BS Legal Management", "BS ME":"BS Management Engineering", "BS MGT":"BS Management",
             "CTM":"BS Communications Technology Management", "ITE":"BS Information Technology Entrepreneurship",
             "LM":"BS Legal Management", "ME":"BS Management Engineering", "MGT":"BS Management",
             "Communications Technology":"BS Communications Technology Management",
             "Information Technology Entrepreneurship":"BS Information Technology Entrepreneurship",
             "Legal Management":"BS Legal Management", "Management Engineering":"BS Management Engineering",
             "Management":"BS Management"} # dictionary to clean

df.loc[:,"course"] = df.loc[:,"course"].apply(lambda course: courseDict[course]) # subject all to Dict

df
````

````python
## phonenumber (part 1)

def numberCleaner(number):
    if number[0] == "9": # 9XX ...
        return re.sub("^9","+63 9",number) # add +63(space)
    else:
        return number

df.loc[:,"phonenumber"] = df.loc[:,"phonenumber"].apply(numberCleaner) # apply to all

df
````

````python
## phonenumber (part 2)

def spaceRemover(number):
    return re.sub("\s","",number) # change space to no space

df.loc[:,"phonenumber"] = df.loc[:,"phonenumber"].apply(spaceRemover) # apply to all
````

````python
df # just checking
````

````python
# carrier

carrierDict = {"+63917":"Globe","+63927":"Globe","+63918":"Smart"}

df["carrier"] = df.loc[:,"phonenumber"].str[0:6].apply(lambda number: carrierDict[number])
df
````

````python
df3 = pd.merge(df, df2.loc[:,["firstname","gender"]], on = "firstname") 
# merge basis firstname
# do not include lastname from df2
````
